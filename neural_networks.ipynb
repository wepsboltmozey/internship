```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
import numpy as np
from tensorflow.keras.layers import Dense, MaxPooling2D, Flatten, Conv2D, Dropout, InputLayer
```

```python
# downloading the data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# normalize the pixel values
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0

# one-hot encoder for feature selection
y_train =to_categorical(y_train, 10)
y_test =to_categorical(y_test, 10)

#add channel dimension for conv2D
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)
```

buiding the model

```python
model = Sequential([
    Conv2D(32, (3, 3), activation ='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2,2)),
    Conv2D(64, (3, 3), activation ='relu'),
    MaxPooling2D((2,2)),
    Conv2D(128, (3, 3), activation ='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation ='softmax'),
])
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs = 20, batch_size =128, validation_split=0.1 )
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test loss: {loss: .4f}, test acuuracy: {accuracy: .4f}')
```

```python
# make predictions
pred = model.predict(x_test[:3])
pred_leb =np.argmax(pred, axis=1)
print('prediction' ,pred_leb)
print('true label:', y_test[:3])
```
